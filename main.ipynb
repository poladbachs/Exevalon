```python
import seaborn as sns
import matplotlib.pyplot as plt

# Adjust the y-axis limit
y_limit = (0, 400)

# Data sizes to plot individually
data_sizes = [100, 1000, 10000]

# Create box plots for each data size
for size in data_sizes:
    # Filter the data for the current data size
    df_filtered = df[df["DataSize"] == size]
    
    # Create the box plot
    ax = sns.catplot(
        data=df_filtered, 
        x="SortingAlgo", 
        y="ExeTime", 
        kind="box",       # Box plot for IQR visualization
        height=6,         # Plot height
        aspect=1.2,       # Adjust width
    )
    
    # Set y-axis limits and labels
    ax.set(ylim=y_limit)
    ax.set_axis_labels("Sorting Algorithm", "Execution Time (ms)")
    
    # Set plot title
    plt.title(f"Execution Time for Data Size {size}")
    
    # Show plot
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a wider violin plot with a higher aspect ratio
ax = sns.catplot(
    data=df, 
    x="SortingAlgo", 
    y="ExeTime", 
    kind="violin", 
    hue="SortingAlgo", 
    split=True,
    height=6,
    aspect=2  # Increase the aspect to make it wider
)

# Set y-axis limit and title
ax.set(ylim=[0, 400])
plt.title("Execution Time Distribution by Sorting Algorithm")
plt.show()
```

```python
ax = sns.catplot(
    data=df, 
    x="SortingAlgo", 
    y="ExeTime", 
    kind="bar", 
    hue="DataSize",
    height=6,
    aspect=1.2
)
ax.set(ylim=[0, 400])
plt.title("Average Execution Time by Sorting Algorithm and Data Size")
plt.show()

```

```python
ax = sns.catplot(
    data=df, 
    x="SortingAlgo", 
    y="ExeTime", 
    kind="swarm", 
    hue="SortingAlgo", 
    height=6,
    aspect=2,
    s=3  # Adjust marker size to make points smaller
)
ax.set(ylim=[0, 400])
plt.title("Execution Time Distribution by Sorting Algorithm")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# List of data sizes to create separate plots for
data_sizes = [100, 1000, 10000]

# Loop through each data size to create individual bar plots
for size in data_sizes:
    # Filter the data for the current data size
    df_filtered = df[df["DataSize"] == size]
    
    # Find the maximum execution time for setting the y-axis limit
    max_exe_time = df_filtered["ExeTime"].max()
    
    # Create the bar plot for the specific data size
    ax = sns.catplot(
        data=df_filtered, 
        x="SortingAlgo", 
        y="ExeTime", 
        kind="bar",        # Bar plot
        height=6,          # Plot height
        aspect=1.5,        # Adjust width
        hue="SortingAlgo"  # Different colors for each sorting algorithm
    )
    
    # Set y-axis limit based on the max execution time for that data size
    ax.set(ylim=(0, max_exe_time * 0.8))  # Adding 10% padding for readability
    ax.set_axis_labels("Sorting Algorithm", "Execution Time (ms)")
    
    # Set plot title for each data size
    plt.title(f"Average Execution Time for Data Size {size}")
    
    # Show plot
    plt.show()
```
